```python
pdf_path = '/content/drive/MyDrive/data-preproccessor/data_preprocessor/data/acetone-acs-l.pdf'
```

```python
from pdf2image import convert_from_path
import os
```

```python
if not os.path.exists('pdf_images'): #directory to store images
    os.makedirs('pdf_images')
```

```python
images = convert_from_path(pdf_path, dpi=300) # convert pdf to images
```

```python
for i, img in enumerate(images):
    img_path = f'pdf_images/page_{i + 1}.png'  # save path for each page
    img.save(img_path, 'PNG')  # save the image as a PNG format
    print(f'Page {i + 1} saved as {img_path}')  # print the saved image path
```

```python
from PIL import Image  #display image
from IPython.display import display
```

```python
for i, img in enumerate(images):  #display each saved image
    display(img)
    print(f'Displaying page {i + 1}')
```

```python
from google.colab import files #download images
```

```python
for i in range(len(images)):    #download each image to local machine
    img_path = f'pdf_images/page_{i + 1}.png'
    files.download(img_path)
```
